```python
from tkinter import *

class Calculadora(object):
    
    def __init__(self, instance):


        self.window = instance
        self.window.title("Calculadora do Marquin v. 0.1")

        self.window.geometry("350x450")

        self.window.configure(bg = "lightblue")

        self.window.wm_iconbitmap("turtle.ico")

        self.font = ("Verdana", "10", "bold")
        self.font2 = ("Verdana", "16", "bold")

        # Frame que contém os checkbuttons
        self.frame1 = Frame(instance, bg = "lightblue", pady = 5)

        # Frame que contém a entrada de texto
        self.frame2 = Frame(instance, bg = "lightblue", pady = 5)

        # Frame que contém o botão calcular
        self.frame3 = Frame(instance, bg = "lightblue", pady = 5)

        # Frame que contém o texto das fórmulas
        self.frame4 = Frame(instance, bg = "lightblue", pady = 5)

        # Frame que contém os botões específicos
        self.frame5 = Frame(instance, bg = "lightblue", pady = 5)

        # Empacotamos as nossas frames
        self.frame1.pack()
        self.frame2.pack()
        self.frame3.pack()
        self.frame4.pack()
        self.frame5.pack()
        

        # Colocar a entrada de texto
        self.form = Entry(self.frame2, font = self.font)
        self.form.pack()

        # Botão calcular
        self.calc = Button(self.frame3, text = "Calcule", command = self.Calcular, font = self.font2)
        self.calc.pack()

        # Texto das fórmulas
        self.result = Label(self.frame4, text = "Resultado", fg = "blue", bg = "lightblue", pady = 10,
                            font = self.font)
        self.result.pack()

        buttons = ("raiz", "/", "*", "-", "+",
                   "7", "8", "9", "%", "1/x",
                   "4", "5", "6", "n!", "x²",
                   "1", "2", "3", "exp(x)", "log(x)",
                   "pi", "0", ".", "<-", "=")
        
        for i in range(len(buttons)):
            if i % 5 == 0:
                subframe = Frame(self.frame5)
                subframe.pack()
            a = Button(subframe, text = buttons[i], bg = "pink",
                       highlightcolor = "red", height = 3, width = 6,
                       font = self.font)
            a.pack(side = LEFT)

    def Calcular(self):

        self.result["text"] = self.form.get()
        self.result["fg"] = "green"
        
        
# Cria a nossa tela
instance = Tk()

# Criamos uma instância da calculadora
Calculadora(instance)

# Dá um título à tela
instance.title("Calculadora do Marquin")
```
