### **Condições**

### Permitem desviar o fluxo lógico de um programa

### **if**: testa se uma determinada condição lógica é verdadeira e executa o comando aninhado ao `if` somente se a condição for verdadeira.

```python
# if

# Uma pessoa pode votar, tirar CNH e se candidatar para presidente da república?
# Depende de sua idade...
idade = float(input("Digite sua idade: "))
print("Isso será sempre impresso na tela, pois está fora do if.")
if idade >= 16:
  print(f"Você já pode votar.")
if idade >= 18:
  print("Já pode tirar sua carteira de habilitação.")
if idade >= 35:
  print("E também já pode se candidatar para presidente da república.")
if idade >= 120:
  print("No entanto, dada a sua idade, provavelmente você já está morto. x_x")
print("Isso também será sempre impresso na tela, pois também está fora do if.")
```

### **if/else**: testa uma condição lógica. Se verdadeiro, executa o comando aninhado ao *statement* `if`; se falso, executa o comando aninhado ao *statement* `else`.

```python
# if/else
num1 = float(input("Digite um número: "))
num2 = float(input("Digite outro número: "))

if num1 > num2:
  print(f"O primeiro número digitado, {num1}, é maior que o segundo número digitado, {num2}.")
else:
  print(f"O primeiro número digitado, {num1}, não é maior que o segundo número digitado, {num2}.")
```

### **elif**: é uma abreviação para `else`/`if`, sendo uma forma de evitar condições aninhadas. O programador pode usar quantos `elif` quiser. Veja:



```python
# Código sem o elif
num1 = float(input("Digite um número: "))
num2 = float(input("Digite outro número: "))

if num1 > num2:
  print(f"O primeiro número digitado, {num1}, é maior que o segundo número digitado, {num2}.")
else:
  if num1 < num2:
    print(f"O primeiro número digitado, {num1}, é menor que o segundo número digitado, {num2}.")
  else:
    print(f"O mesmo número, {num1}, foi digitado duas vezes.")
```

```python
# Mesma solução anterior, mas agora evitando o else/if aninhados
# Código sem o elif
num1 = float(input("Digite um número: "))
num2 = float(input("Digite outro número: "))

if num1 > num2:
  print(f"O primeiro número digitado, {num1}, é maior que o segundo número digitado, {num2}.")
elif num1 < num2:
  print(f"O primeiro número digitado, {num1}, é menor que o segundo número digitado, {num2}.")
else:
  print(f"O mesmo número, {num1}, foi digitado duas vezes.")
```

### **Exercício 1**: Escreva um programa que recebe um número e que informa se ele é divisível por 2, por 3, por 5 e por 7.

```python
num = float(input("Digite um número aí, manolito: "))

if (num % 2 == 0):
  print(f"{num} é divisível por 2.")
else:
  print(f"{num} não é divisível por 2.")
if (num % 3 == 0):
  print(f"{num} é divisível por 3.")
else:
  print(f"{num} não é divisível por 3.")
if (num % 5 == 0):
  print(f"{num} é divisível por 5.")
else:
  print(f"{num} não é divisível por 5.")
if (num % 7 == 0):
  print(f"{num} é divisível por 7.")
else:
  print(f"{num} não é divisível por 7.")
```

### **Exercício 3**: Escreva um programa que calcule o número de salgados comprados por um cliente. Se ele comprou mais de 5, dê 3% de desconto a ele; se ele comprou mais de 10, dê 5% de desconto a ele; mais de 20, dê 10% de desconto a ele; a partir de 30, dê 15% de desconto a ele. O valor de cada salgado é 5 reais.

```python
salgados = int(input("Digite o número de salgados que você quer comprar? "))

if salgados < 5:
  preco = salgados * 5
elif 5 <= salgados < 10:
  preco = salgados * 5 * 0.97
elif 10 <= salgados < 20:
  preco = salgados * 5 * 0.95
elif 20 <= salgados < 30:
  preco = salgados * 5 * 0.90
else:
  preco = salgados * 5 * 0.85

print(f"Você comprou {salgados} salgados e pagará R$ {preco:.2f}.")
```

### **Exercício 4**: Escreva um programa que recebe três notas de um aluno e retorne sua nota e o seguinte resultado: (a) Aprovado: média >= 7; (b) Recuperação: média > 5 e < 7; e (c) Reprovado: média < 5.

```python
nota1 = float(input("Digite uma nota entre 0 e 10: "))
nota2 = float(input("Digite uma nota entre 0 e 10: "))
nota3 = float(input("Digite uma nota entre 0 e 10: "))

media = (nota1 + nota2 + nota3) / 3

if media >= 7:
  print(f"Parabéns! Sua média final foi {media:.2f} e você foi APROVADO!")
elif 5 <= media < 7:
  print(f"Deu mole! Sua média final foi {media:.2f} e você ficou de RECUPERAÇÃO.")
else:
  print(f"Você foi muito cabaço. Sua média final foi {media:.2f} e você foi REPROVADO!")
```
