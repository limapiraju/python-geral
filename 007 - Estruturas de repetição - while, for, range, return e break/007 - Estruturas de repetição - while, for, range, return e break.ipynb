### **while**: executa um laço de repetição enquanto uma determinada condição for verdadeira.

```python
# Exemplo 1: Imprime números de 0 a 10
numero = 0
while numero <= 10:
  print(numero, end = " ")
  # incrementa o valor de numero em uma unidade para evitar um loop infinito!
  numero += 1
```

```python
# Exemplo 2: Contagem regressiva
print("Atenção para a contagem regressiva...")
numero = 10
while numero >= 0:
  print(f"{numero}...")
  # decrementa o valor de numero em uma unidade para evitar um loop infinito!
  numero -= 1
```

### Um `while` *loop* pode ser já iniciado como verdadeiro. Neste caso, é preciso inserir alguma maneira dele ser quebrado internamente. Isso é feito pelo *statement* `break`, que quebra o *loop* quando uma condição se torna verdadeira.

```python
# Exemplo 3: Qual é a resposta fundamental da vida, do Universo e de tudo mais?
resposta = 0
while True:
  if resposta == 42:
    print(f"{resposta} é a resposta!!!")
    break

  print(f"{resposta} definitivamente NÃO É a resposta...")
  resposta += 1
```

### De maneira alternativa, podemos querer que uma iteração do *loop* seja interrompida e que prossigamos para  uma nova iteração quando alguma condição é ou não é atendida. Obtemos isso por meio do *statement* `continue`.

```python
# Exemplo 4: Qual é a resposta fundamental da vida, do Universo e de tudo mais?
resposta = 0
while True:
  if resposta != 42:
    resposta += 1
    continue
  else:
    print(f"{resposta} é a resposta!!!")
    break
```

```python
# Exemplo 5: Recebe números do usuário até ele digitar 0. Retorna a soma de todos os números pares.
soma = 0
n = 0

while True:
  num = int(input("Digite um número inteiro (ou digite 0 para sair do programa.): "))
  if num == 0:
    break
  # se o número digitado for par
  if num % 2 == 0:
    soma += num
  n += 1
  continue

print(f"Você digitou {n} números, cuja soma dos números pares foi igual a {soma}.")
```

### **for**: enquanto o `while` percorre um *loop* de tamanho indeterminado, o `for` percorre um *loop* de tamanho determinado. Para isso, ele é associado so *statement* `range`, que retorna uma sequência iterável.

```python
# Exemplo 6: Imprime os números pares de 2 a 100

# do número i, sendo i assumindo valores entre 2 (inclusive) e 102 (exclusive), a passos de 2 em 2:
for i in range(2, 102, 2):
  if (i -2) % 10 == 0:
    print("\n")
  print(i, end = "\t")
```

```python
# Exemplo 7: Imprime os números pares de 2 a 100

# realiza a mesma operação anterior, mas de outra maneira
for i in range(1, 101):
  if i % 2 == 1:
    continue
  if (i -2) % 10 == 0:
    print("\n")
  print(i, end = "\t")
```

### ***Loops* aninhados**: consiste em encaixar um *loop* dentro do outro. Veja:

```python
# Exemplo 8: imprime a tabuada do 2 ao 10

# loop externo: o primeiro fator da tabulada
for i in range(2, 11):
  print(f"Tabulada do {i}:")
  # loop interno: o segundo fator da tabuada, que varia de 0 a 10
  for j in range(0, 11):
    print(f"{i:<2} × {j:>2} = {i*j:>3}")
  print("\n")
```
